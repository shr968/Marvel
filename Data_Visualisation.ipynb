```python
# Import necessary libraries
import pandas as pd
import plotly.express as px
data = {
    'world_rank': range(1, 101),
    'university_name': [f'University {i}' for i in range(1, 101)],
    'citations': [1000 + i * 10 for i in range(100)],
    'teaching': [500 + i * 5 for i in range(100)]
}
timesData = pd.DataFrame(data)

fig = px.scatter(
    timesData,
    x='world_rank',
    y='citations',
    size='teaching',
    color='university_name',
    hover_name='university_name',
    title='Citations vs World Rank of Top 100 Universities',
    labels={'world_rank': 'World Rank', 'citations': 'Citations'}
)

fig.show()
```
