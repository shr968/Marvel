```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(42)
data = {
    'A': np.random.randint(1, 20, 10),
    'B': np.random.randint(1, 20, 10),
    'C': np.random.randint(1, 20, 10),
    'Category': np.random.choice(['Group 1', 'Group 2'], 10)
}
df = pd.DataFrame(data)

def create_plots():
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle("Page 1")

    axes[0, 0].plot(df['A'], label="Line Plot")
    axes[0, 0].set_title("Line Plot")
    axes[0, 0].set_xlabel("Index")
    axes[0, 0].set_ylabel("Value")
    axes[0, 0].legend()

    axes[0, 1].fill_between(df.index, df['A'], color="skyblue", alpha=0.4)
    axes[0, 1].plot(df['A'], color="Slateblue", alpha=0.6, linewidth=2)
    axes[0, 1].set_title("Area Plot")
    axes[0, 1].set_xlabel("Index")
    axes[0, 1].set_ylabel("Value")

    axes[1, 0].scatter(df['A'], df['B'], label="Scatter Plot")
    axes[1, 0].set_title("Scatter Plot")
    axes[1, 0].set_xlabel("A")
    axes[1, 0].set_ylabel("B")
    axes[1, 0].legend()

    axes[1, 1].scatter(df['A'], df['B'], s=df['C']*10, alpha=0.5, color='green', label='Bubble')
    axes[1, 1].set_title("Bubble Plot")
    axes[1, 1].set_xlabel("A")
    axes[1, 1].set_ylabel("B")
    axes[1, 1].legend()

    plt.show()

    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle("Page 2")

    axes[0, 0].bar(df.index, df['A'], label="A")
    axes[0, 0].set_title("Simple Bar Plot")
    axes[0, 0].set_xlabel("Index")
    axes[0, 0].set_ylabel("A")
    axes[0, 0].legend()

    axes[0, 1].bar(df.index - 0.2, df['A'], width=0.4, label="A")
    axes[0, 1].bar(df.index + 0.2, df['B'], width=0.4, label="B")
    axes[0, 1].set_title("Grouped Bar Plot")
    axes[0, 1].set_xlabel("Index")
    axes[0, 1].set_ylabel("Values")
    axes[0, 1].legend()

    axes[1, 0].bar(df.index, df['A'], label="A")
    axes[1, 0].bar(df.index, df['B'], bottom=df['A'], label="B")
    axes[1, 0].set_title("Stacked Bar Plot")
    axes[1, 0].set_xlabel("Index")
    axes[1, 0].set_ylabel("Values")
    axes[1, 0].legend()

    axes[1, 1].hist(df['A'], bins=5, color="purple", alpha=0.7)
    axes[1, 1].set_title("Histogram")
    axes[1, 1].set_xlabel("A")
    axes[1, 1].set_ylabel("Frequency")

    plt.show()

    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle("Page 3")

    axes[0, 0].pie(df['A'], labels=df.index, autopct='%1.1f%%', startangle=140)
    axes[0, 0].set_title("Pie Plot")

    sns.boxplot(data=df[['A', 'B', 'C']], ax=axes[0, 1])
    axes[0, 1].set_title("Box Plot")

    sns.violinplot(data=df[['A', 'B', 'C']], ax=axes[1, 0])
    axes[1, 0].set_title("Violin Plot")

    X, Y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
    Z = np.sin(np.sqrt(X**2 + Y**2))
    contour = axes[1, 1].contourf(X, Y, Z, cmap='coolwarm')
    fig.colorbar(contour, ax=axes[1, 1])
    axes[1, 1].set_title("Contour Plot")

    plt.show()

    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    sns.heatmap(pd.DataFrame(np.random.randn(10, 10)), cmap="viridis", ax=ax)
    ax.set_title("Heatmap")
    plt.show()

    sns.pairplot(df[['A', 'B', 'C', 'Category']], hue='Category', palette="Set2")
    plt.suptitle("Multivariate Distribution Plot", y=1.02)
    plt.show()

create_plots()
```
